# 월별 거래량 그래프

```python
p=p_pro_month_count['count(CLAC1_NM)']
p1=list()
for i in [5,4,3,2,1,0]:{
    p1.append(p[i])
}   
    
    
month = ["18.04", "18.05", "18.06", "18.07", "18.08", "18.09"]
rev_month = ["18.09","18.08", "18.07", "18.06", "18.05","18.04"]

plt.fig, ax = plt.subplots(1,1,figsize=(10,4))

plt.bar = ax.barh(rev_month,p1 , label = "거래량", color = "#2F4554")

plt.title("월별 거래량", size = 20)
plt.ylabel("월", size = 15)
plt.xlabel("총 구매수량 (개)", size = 15)
plt.show()
```
