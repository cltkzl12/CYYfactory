```python
from pykrx import stock
tickers = stock.get_index_ohlcv_by_date('20190101','20200727','코스피')
print(tickers)
```

```python
tickers = tickers.reset_index()
```

```python
tickers.head()
```

```python
# 날짜 인덱싱
copy_ticker = tickers.copy()

for i in range(len(tickers['날짜'])):
    copy_ticker.loc[i,'날짜'] = str(tickers.loc[i]['날짜'])[0:7]

# copy_ticker.loc[0,'날짜'] =str(ticker.loc[0]['날짜'])[0:7]
copy_ticker
```

```python
copy_ticker = copy_ticker.groupby('날짜').min().reset_index()
```

```python
def grap_year(data):
    data = str(data)
    return int(data[:4])

def grap_month(data):
    data = str(data)
    return int(data[5:])
# 날짜 처리

copy_ticker['year'] = copy_ticker['날짜'].apply(lambda x: grap_year(x))
copy_ticker['month'] = copy_ticker['날짜'].apply(lambda x: grap_month(x))
copy_ticker = copy_ticker.drop(['날짜'], axis=1)
copy_ticker.head()
```

```python
copy_ticker.to_csv('kospi_stock.csv',encoding='utf-8')
```
