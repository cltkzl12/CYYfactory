### 크기조정 : cv2.resize()

```python
img = cv2.imread('image.jpg')
r,c = img.shape[:2]
print('row:',r,'col:',c)
```

```python
# 이미지 2배 확대
new_image = cv2.resize(img, (2*r,2*c), interpolation=cv2.INTER_CUBIC)
cv2.imwrite('resize2_image.jpg',new_image)
cv2.imshow('resize_image',new_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 이미지 1/2 축소
new_image = cv2.resize(img,None,fx=0.5,fy=0.5, interpolation=cv2.INTER_AREA)
cv2.imshow('resize_image',new_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 확대(1<fx) or 축소(1>fx) 모두 가능
new_image = cv2.resize(img,None,fx=1.5,fy=1.5, interpolation=cv2.INTER_LINEAR)
cv2.imshow('resize_image',new_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 이미지 자르기
crop_image = img[10:200,130:330]
cv2.imshow('crop_image', crop_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 회전
r,c = img.shape[:2]
M = cv2.getRotationMatrix2D((c/2,r/2),90,1) # 중심 좌표,회전각(CCW,시계반대방향),크기비율
new_image = cv2.warpAffine(img,M,(c,r))
cv2.imshow('rotation_image', new_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 임계값 처리 : threshold
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
new_image = cv2.threshold(gray_img,120,255,cv2.THRESH_BINARY)
# 120 보다 큰 값은 흰색으로 그외에는 검은색으로 출력
# 값이 작을수록 흰색이 많이 출력
# 값이 클수록 검은색이 많이 출력
cv2.imshow('threshold_image', new_image[1])
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 필터 : filter2D(),합성곱
ker = np.array([[1,1,1],
                [1,1,1],
                [1,1,1]]) # (3,3)
new_image = cv2.filter2D(img,-1,ker)
cv2.imshow('filter2D_image', new_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 가우시안 블러 : GaussianBlur()
new_image = cv2.GaussianBlur(img,(5,5),0) # (5,5) : 반드시 홀수,클수록 흐림
cv2.imshow('gaussian_image', new_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 중간값 블러 : medianBlur()
new_image = cv2.medianBlur(img,11) # 반드시 홀수,클수록 흐림
cv2.imshow('medain_image', new_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 침식: erode
# 두 물체를 축소시켜서 두 물체의 차이가 명확해지게 하거나, 노이즈를 제거
img = cv2.imread('image.jpg')  # 512*512*3,(512,512,3)
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow("original_image", gray_img)
ker = np.ones((5,5),np.uint8)
new_img = cv2.erode(gray_img,ker,iterations = 1)
cv2.imshow("erosion_image", new_img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 팽창 : dilate()
# 작은부분을 확대하려는 경우, 원하지 않는 틈/구멍을 메우고 싶을때
img = cv2.imread('image.jpg')  # 512*512*3,(512,512,3)
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow("original_image", gray_img)
ker = np.ones((5, 5), np.uint8)
new_img = cv2.dilate(gray_img, ker, iterations=1)
cv2.imshow("dilation_image", new_img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 소벨 에지검출 : Sobel()
img = cv2.imread('image.jpg')  # 512*512*3,(512,512,3)
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow("original_image", gray_img)
x_edges = cv2.Sobel(gray_img,-1,1,0,ksize = 5) # 1,0 : x축방향
y_edges = cv2.Sobel(gray_img,-1,0,1,ksize = 5) # 0,1 : y축방향
cv2.imshow("x_edges_image", x_edges)
cv2.imshow("y_edges_image", y_edges)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 캐니 에지검출: Canny()
img = cv2.imread('image.jpg')  # 512*512*3,(512,512,3)
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow("original_image", gray_img)
edges = cv2.Canny(gray_img, 100,200,5) # min,max,aperture_size(=kernel_size)
cv2.imshow("canny_image", edges)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 윤곽선 검출 : findContours(), drawContours()
img = cv2.imread('image.jpg')  # 512*512*3,(512,512,3)
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
thresh_image = cv2.threshold(gray_img, 127, 255, cv2.THRESH_BINARY)
contours,hierarchy = cv2.findContours(thresh_image[1],cv2.RETR_TREE,\
                                       cv2.CHAIN_APPROX_SIMPLE  )
cv2.drawContours(img,contours,-1,(255,0,0),2) # (B,G,R):윤곽선색상 ,  2: thickness
cv2.imshow("contour_image", img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# template_image 이미지 생성
img = cv2.imread('image.jpg')
crop_image = img[10:200,130:330]
cv2.imwrite('template_image.jpg',crop_image)
cv2.imshow('crop_image', crop_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

### 동영상 프레임에 사각형과 문자 넣기

```python
import cv2

# 영상 정보 가져오기
video_input_path = 'InfiniteWar01.mp4'
video_output_path = 'InfiniteWar01_rectangle_out.mp4'  # (Linux 에서는 반드시 .avi)

cap = cv2.VideoCapture(video_input_path)  # cv2.VideoCapture(0) : 웹캠 사용

# Codec은 *'XVID'로 설정. 
codec = cv2.VideoWriter_fourcc(*'XVID')

# 영상 정보
vid_size = (round(cap.get(cv2.CAP_PROP_FRAME_WIDTH)),round(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))) # (1216, 676)
vid_fps = cap.get(cv2.CAP_PROP_FPS)
frame_cnt = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

vid_writer = cv2.VideoWriter(video_output_path, codec, vid_fps, vid_size)
print('총 Frame 갯수:', frame_cnt, '  FPS:', round(vid_fps), '  Frame 크기:', vid_size)
```

```python
green_color = (0,255,0)
red_color = (0,0,255)

import time

index = 0
while True:
    hasFrame, img_frame = cap.read()
    if not hasFrame:
        break
    index +=1    
    print('frame : ',index, '처리 완료')
    
    # 사각형 그리기
    cv2.rectangle(img_frame,(200,150,800,400), color=green_color, thickness=2)
    
    # 문자 넣기
    caption = 'rectangle_image frame: {}'.format(index)
    cv2.putText(img_frame, caption, (200,145), cv2.FONT_HERSHEY_SIMPLEX, 0.7, red_color, 1)
    
    # 사각형과 문자를 넣어 변환된 이미지 출력
    cv2.imshow("original_image", img_frame)
    
    vid_writer.write(img_frame)

    # 캐니 에지 검출: Canny()
    img_frame = cv2.cvtColor(img_frame, cv2.COLOR_BGR2GRAY)
    img_frame = cv2.Canny(img_frame, 100,200,5) # min,max,aperture_size(=kernel_size)

    # 캐니 변환된 이미지 출력
    cv2.imshow('InfinitiWar',img_frame)
    
    time.sleep(0.03)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
cv2.destroyAllWindows()
vid_writer.release()
cap.release()
```

### 동영상 프레임의 얼굴과 눈 인식 표시하기

```python
import cv2

# 영상 정보 가져오기
video_input_path = 'InfiniteWar01.mp4'
video_output_path = 'InfiniteWar01_face_eye_out.mp4'  # (Linux 에서는 반드시 .avi)

cap = cv2.VideoCapture(video_input_path)  # cv2.VideoCapture(0) : 웹캠 사용

# Codec은 *'XVID'로 설정. 
codec = cv2.VideoWriter_fourcc(*'XVID')

# 영상 정보
vid_size = (round(cap.get(cv2.CAP_PROP_FRAME_WIDTH)),round(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))) # (1216, 676)
vid_fps = cap.get(cv2.CAP_PROP_FPS)
frame_cnt = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

vid_writer = cv2.VideoWriter(video_output_path, codec, vid_fps, vid_size)
print('총 Frame 갯수:', frame_cnt, '  FPS:', round(vid_fps), '  Frame 크기:', vid_size)
```

```python
def detect_face_eye(img):
    face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
    eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(img_gray, 1.3, 5)
    for (x, y, w, h) in faces:
        cv2.rectangle(img, (x, y), (x + w, y + h), (0, 0, 255), 2)
        roi_gray = img_gray[y:y + h, x:x + w]
        roi_color = img[y:y + h, x:x + w]
        eyes = eye_cascade.detectMultiScale(roi_gray)
        for (ex, ey, ew, eh) in eyes:
            cv2.rectangle(roi_color, (ex, ey), (ex + ew, ey + eh), (0, 255, 0), 2)
    cv2.imshow("face_eye_detect_image", img)
    return img

# 이미지로 테스트
img = cv2.imread("image.jpg")  
detect_face_eye(img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
green_color = (0,255,0)
red_color = (0,0,255)

import time

index = 0
while True:
    hasFrame, img_frame = cap.read()
    if not hasFrame:
        break
    index +=1    
    print('frame : ',index, '처리 완료')
    
    # 문자 넣기
    caption = 'detect_face_eye frame: {}'.format(index)
    cv2.putText(img_frame, caption, (200,145), cv2.FONT_HERSHEY_SIMPLEX, 0.7, red_color, 1)
    
    # face와 eye를 검출하여 사각형 표시
    img_frame = detect_face_eye(img_frame)
    
    # 변환된 이미지를 저장
    vid_writer.write(img_frame)

    # time.sleep(0.03)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
cv2.destroyAllWindows()
vid_writer.release()
cap.release()
```
