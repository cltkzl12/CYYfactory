```python
def grap_year(data):
    data = str(data)
    return int(data[:4])

def grap_month(data):
    data = str(data)
    return int(data[4:])
```

```python
# 날짜 처리
data = pd.read_csv('202004.csv')
data = data.fillna('')
data['year'] = data['REG_YYMM'].apply(lambda x: grap_year(x))
data['month'] = data['REG_YYMM'].apply(lambda x: grap_month(x))
data = data.drop(['REG_YYMM'], axis=1)
```

```python
# 데이터 정제
df = data.copy()
df = df.drop(['CARD_CCG_NM', 'HOM_CCG_NM'], axis=1)

columns = ['CARD_SIDO_NM', 'STD_CLSS_NM', 'HOM_SIDO_NM', 'AGE', 'SEX_CTGO_CD', 'FLC', 'year', 'month']
df = df.groupby(columns).sum().reset_index(drop=False)
```

```python
pre_df = df.drop(['AMT'],axis=1)
pre_df_cp = pre_df.copy()
pre_df_cp['month'] =7
# pre_df = pre_df[(pre_df['year']==2020) & (pre_df['month']==3)]
# pre_df = pre_df[(pre_df['year']==2020) & (pre_df['month']==3)]
# pre_df['month']=4
pre_df = pre_df.append(pre_df_cp)
```
